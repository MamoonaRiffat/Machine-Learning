### Step 17. What is the mean age of users?

```python
sum_ages=user_id.age.sum()
print(sum_ages)
```

```python
mean_age=sum_ages/943     # there are 943 rows, so total users are 943
print("Mean age of the users is:",np.round(mean_age,2))
```

### Step 18. What is the age with least occurrence?

```python
value_counts=user_id['age'].value_counts()  ## counts of occurences of each value
least_occ=value_counts.min()
print("The least occurence age is:",np.round(least_occ,2))
```
