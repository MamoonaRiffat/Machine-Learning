```python
# logistic regression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
student_d=pd.read_csv("student_data.csv") # loading file from computer
student_d.sample(10)
 
```

```python
student_d = student_d.drop(['lunch'],axis=1)
student_d = student_d.drop(['parental level of education'],axis=1)
student_d = student_d.drop(['race/ethnicity'],axis=1)
student_d = student_d.drop(['test preparation course'],axis=1)
student_d.sample(10)
```

```python
# label encode 0 and 1; 0 for male and 1 for female in gender column

from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Fit and transform the 'Category' column
student_d['encoded_gender'] = label_encoder.fit_transform(student_d['gender'])
```

```python
student_d.head()   #female 0 and male 1
```

```python
#feature and target and drop sex column


student_d.drop('gender', axis=1, inplace=True)
```

```python
x=student_d.drop(columns='encoded_gender')
y=student_d['encoded_gender']
```

```python
#spliiting
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#model 
lr = LogisticRegression()
#train
lr.fit(X_train, y_train)
```

```python
#TEST
y_pred = lr.predict(X_test)
```

```python
#evaluation
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Generate a classification report
class_report = classification_report(y_test, y_pred)
print("Classification Report:\n", class_report)

# Generate a confusion matrix
confusion_mat = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", confusion_mat)
```
