Task: 2,2 graph using seaborn and matplotlib

```python
#Plot1:  MATPLOT LIB
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

```

```python
data = np.random.randn(100000)  ## generate random data 
```

```python
# Creating a histogram graph using matplotlib
plt.hist(data, color='pink')
plt.title('Histogram')
plt.xlabel('X-axis Label')
plt.ylabel('Frequency')
plt.show()
```

##Plot 2: MATPLOT LIB

## Pie Chart using matplotlib: 

```python
# Data for the pie chart
labels = ['Natural gas', 'Fuel', 'Hydrogen Gas', 'Diesel']
sizes = [25, 30, 15, 20]  # These values represent the proportions (percentages) of each category fuels
plt.figure(figsize=(6, 6))  #(height, width)
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Fuel Proportions')
plt.show()
```

TASK : Histogram graph using Seaborn

```python
# Generating random data
data = np.random.randn(100000)

# Creating a histogram using Seaborn
sns.histplot(data,bins=40, color='skyblue')
plt.xlabel('X-axis')
plt.ylabel('Frequency')
plt.title('Histogram with Seaborn')
plt.show()
```

Task: Grouped boxplots

```python

data = pd.DataFrame({
    'Group': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'], # students of different sections in a college
    'students': [10, 15, 12, 5, 8, 7, 20, 18, 22]   #no of students
})

# grouped boxplot using Seaborn
plt.figure(figsize=(8, 6))
sns.boxplot(x='Group', y='students', data=data, palette='Set2')
plt.xlabel('Group')
plt.ylabel('students')
plt.title('Grouped Boxplot')
plt.show()
```
